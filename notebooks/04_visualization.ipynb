```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from config import engine

df_orders_pay = pd.read_sql("SELECT * FROM order_payments", engine) #df eh dataframe
plt.figure(figsize=(10,6))
plt.title("Distribuição de valores")
sns.histplot(df_orders_pay['payment_value'], bins = 100, kde=True) #a linha kde indica estimativa da densidade de dados, onde são mais frequentes
```

```python

df_products = pd.read_sql("SELECT * FROM products", engine) #df eh dataframe
plt.figure(figsize=(10,6))
plt.title("Distribuição de valores")
sns.histplot(df_products['product_weight_g'], bins = 10, kde=True)
```

```python
df_order_items = pd.read_sql("SELECT * FROM order_items", engine) #df eh dataframe
plt.figure(figsize=(10,6))
plt.title("Distribuição de valores")
sns.scatterplot(x = df_order_items['price'], y=df_order_items['product_id'])
plt.show()
```
